```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001,1002,1003,1004,1005,1006,1007],
    '이름' : ['둘리','도우너','또치','길동','희동','마이콜','영희']},
    columns=['고객번호','이름'])
df1
```

```python
df11 = pd.DataFrame({
    '고객번호': [1009,1010,1011,1012,1013,1014,1007],
    '이름' : ['둘리','도우너','또치','길동','희동','마이콜','영희동생']},
    columns=['고객번호','이름'])
df11
```

```python
df111 = pd.merge(df1,df11, how='outer')
df111
```

```python
df11111 = pd.concat([df1,df11])
df11111
```

```python
df111111 = pd.concat([df1,df11], ignore_index=True)
df111111
```
