### 4. Data Learning

#### 1) 종속변수 독립변수 추출

```python
X=np.array(df.drop(columns='processState_A')) #종속변수
Y=np.array(df.processState_A) #독립변수
```

```python
table = itemfreq(Y)
x_ticks = ['0','1'] # 0 : 입양X, 1:입양O
plt.bar(x_ticks, table[:,1], color='pink')
plt.show()
table
```

#### 2) 데이터셋 나누기(train/test) 

```python
X_train, X_test, Y_train, Y_test = train_test_split(
                                    X,Y,test_size=0.3, random_state=5)
# train_test_split
# : 데이터를  train set과 test set으로 단순분리
# : 원래 트레이닝 데이터와 결과값이 X와 Y를 무작위로 섞은 후 test_size로 주어진 비율만큼 나눔
# test_size 0.3은 X와 Y를 무작위로 섞은 후에 70%은 train으로, 30%는 test로 하는것
# random_state : 난수발생을 위한 seed의 인자값
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

#### 3) 모형 적용 : NaiveBayes

```python
GNB = GaussianNB()
GNB.fit(X_train, Y_train) #훈련
Y_pred_test = GNB.predict(X_test)
Y_pred_test
```

#### 4) 검증

```python
# 혼동행렬
# : 입력에 대해 모델이 예측한 결과 클래스와 실제 목표 클래스가 
# : 일치하는 개수를 센 결과를 반환
conf_mat = metrics.confusion_matrix(Y_test,Y_pred_test)
conf_mat
```

```python
np.sum(conf_mat)
```

```python
accuracy = metrics.accuracy_score(Y_test, Y_pred_test)
print('Accuracy    = ' + str(np.round(accuracy,2)))
```
